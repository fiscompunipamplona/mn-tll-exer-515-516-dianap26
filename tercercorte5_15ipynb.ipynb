```


def f(x):
  return 1 + 0.5 * tanh(2*x)

#calcular df/dx utilizando el metodo de diferencia central
def df_dx(x):
  h = 10 ** -5  # numero de pie
  return (f(x + 0.5 * h) - f(x - 0.5 * h)) / h
def g(x):
  #derivada analitica de f(x) anterior
  return 1 / cosh(2*x) ** 2
xvals = linspace(-2, 2, 100)
dfvals = list(map(df_dx, xvals))
gvals = list(map(g, xvals))
plot(xvals, dfvals, 'o')
plot(xvals, gvals)
xlabel('x')
show()

```
